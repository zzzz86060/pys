```python
# 首先通过使用librosa.load导入librosa内置示例音频文件
import librosa
import librosa.display
import matplotlib.pyplot as plt
y, sr = librosa.load(librosa.example("choice"),duration=10)
# 绘制时域波形图
plt.figure()
# 颜色需要指明，不然会报错
librosa.display.waveshow(y,sr=sr,color="blue")
plt.show()
```

```python
melspec = librosa.feature.melspectrogram(y=y, sr=sr)
logmelspec = librosa.power_to_db(melspec)
plt.figure()
librosa.display.specshow(logmelspec, sr=sr, x_axis="time", y_axis="mel")
plt.show()
```

```python
y, sr = librosa.load(librosa.example("choice"),duration=10)
rms = librosa.feature.rms(y=y)
print(rms.shape)
```

```python
plt.figure()
plt.semilogy(rms.T,label="RMS Energy")
plt.xticks([])
plt.xlim([0,rms.shape[-1]])
plt.legend()
plt.tight_layout()
plt.show()
```

```python
y,sr = librosa.load(librosa.example("robin"),duration=10)
cent = librosa.feature.spectral_centroid(y=y,sr=sr)
print(cent.shape)
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.semilogy(cent.T,label="Spectral centroid")
plt.ylabel("Hz")
plt.xticks([])
plt.xlim([0,cent.shape[-1]])
plt.legend()
plt.show()
```

```python
y,sr = librosa.load(librosa.example("vibeace"), duration=10)
spec_bw = librosa.feature.spectral_bandwidth(y=y,sr=sr)
print(spec_bw.shape)
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.semilogy(spec_bw.T,label="Spectral centroid")
plt.ylabel("Hz")
plt.xticks([])
plt.xlim([0,spec_bw.shape[-1]])
plt.legend()
plt.show()
```

```python
y,sr = librosa.load(librosa.example("vibeace"), offset=30, duration=5)
mfcc_feature = librosa.feature.mfcc(y=y,sr=sr)
print(mfcc_feature.shape)
```

```python
plt.figure(figsize=(10,4))
librosa.display.specshow(mfcc_feature, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```
